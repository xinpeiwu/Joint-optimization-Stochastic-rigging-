```python
import numpy as np
from pyDOE import lhs

# 定義整數變數範圍
integer_bounds = [(1, 2), (1, 2), (1, 2), (1, 2)]  # 整數變數範圍
float_bounds = [(0, 35), (0, 35), (0, 35), (0, 35), (0, 38), (0, 38), (0, 38), (0, 38), (0, 41), (0, 41),
                (0, 41), (0, 41), (0, 45), (0, 45), (0, 45), (0, 45), (1, 5), (10, 15), (2, 6), (11, 17),
                (2, 6), (12, 18), (3, 8), (13, 20)]  # Float variable ranges

# 希望生成的樣本數量
num_samples = 100

# 初始化一個空的數組，用於存儲整數 LHS 抽樣結果
lhs_samples_integer = np.zeros((num_samples, len(integer_bounds)), dtype=int)

# 進行100次整數 LHS 抽樣
for i in range(num_samples):
    lhs_samples_int = lhs(len(integer_bounds), samples=1)
    for j in range(len(integer_bounds)):
        lower_bound, upper_bound = integer_bounds[j]
        lhs_samples_int[:, j] = np.round(lhs_samples_int[:, j] * (upper_bound - lower_bound) + lower_bound).astype(int)
    lhs_samples_integer[i] = lhs_samples_int

# 進行浮點數 LHS 抽樣
lhs_samples_float = lhs(len(float_bounds), samples=num_samples)
for i in range(len(float_bounds)):
    lower_bound, upper_bound = float_bounds[i]
    lhs_samples_float[:, i] = lower_bound + lhs_samples_float[:, i] * (upper_bound - lower_bound)
    lhs_samples_float[:, i] = np.round(lhs_samples_float[:, i], decimals=2)

# 合併整數和浮點數 LHS 抽樣結果
lhs_samples = np.concatenate((lhs_samples_integer, lhs_samples_float), axis=1)

print(lhs_samples)
```
